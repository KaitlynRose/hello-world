```python
n=4
def fact(n):
    if n==1:
        return 1
    else:
        return n*fact(n-1)
fact(n)
```

```python
import turtle
import numpy as np
bob=turtle.Turtle()
wn=turtle.Screen()

def koch(t,length):
    if length<= 10:
        t.fd(length)
        return
    m=length/3
    koch(t,m)
    t.lt(60)
    koch(t,m)
    t.rt(120)
    koch(t,m)
    t.lt(60)
    koch(t,m)



def dragon(t,length):
    if length <= 0:
        t.fd(length)
        return
    m=length/np.sqrt(2)
    t.lt(45)
    dragon(t,m)
    t.rt(90)
    dragon(t,m)
    t.lt(45)

dragon(bob,100)

#koch(bob,100)
#bob.rt(120)
#koch(bob,100)
#bob.rt(120)
#koch(bob,100)

wn.exitonclick()
turtle.mainloop()
```
