```python
import random
import turtle

def get_square(t,sz):
    #make a square sz by sz
    for i in range(4):
        t.forward(sz)
        t.left(90)


#global variables







#set up a turtle to play
wn =turtle.Screen()
wn.bgcolor("lightgreen")

flame = turtle.Turtle()
get_square(flame,50)


r = random.randrange(0, 257, 10)
g = random.randrange(0, 257, 10)
b = random.randrange(0, 257, 10)


def drawRectangle(t, w, h):
    wn.colormode(255)
    x=random.randrange(-300, 300)
    y=random.randrange(-250, 250)
    t.setx(x)
    t.sety(y)
    t.color(r, g, b)
    t.begin_fill()
    for i in range(2):
        t.forward(w)
        t.right(90)
        t.forward(h)
        t.right(90)
    t.end_fill()
    t.penup()
drawRectangle(flame,40,50)
```
